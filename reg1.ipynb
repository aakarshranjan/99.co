```python

import pandas as pd


# reading the data
df = pd.read_csv('D:/private-residential-property-price-index-base-quarter-2009-q1-100.csv')
```

```python
print(df)
```

```python
#DROPING THE QUARTER AND LEVEL_1 COLUMN
df=df.drop(['quarter'],1)
```

```python
df=df.drop(['level_1'],1)
```

```python
print(df)
```

```python
#CREATING A NEW PREDICTION COLUMN
df['prediction']=df[['value']].shift(-1)
```

```python
print(df)
```

```python
#DROPPPING THE LAST ROW
df.drop(df.index[177])
```

```python

#CONVERTING X INTO NUMPY ARRAY
import numpy as np
x=np.array(df.value)
```

```python
print(x)
```

```python
x.shape
```

```python
x=np.delete(x,177,0)
```

```python
#CONVERTING y INTO NUMPY ARRAY
y=np.array(df['prediction'])
```

```python
y=np.delete(y,177,0)
```

```python
print(x.shape)
```

```python
print(y.shape)
```

```python

#CREATING THE TRAINING AND TESTING DATAS USING TRAIN_TEST_SPLIT
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.2)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
X_train=X_train.reshape(-1,1)
print(X_train)
```

```python
y_train=y_train.reshape(-1,1)
```

```python
print(y_train)
```

```python
#USING LINEAR REGRESSION AS A MODEL
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
X_test=X_test.reshape(-1,1)
print(X_test)
```

```python
y_test=y_test.reshape(-1,1)
print(y_test)
```

```python
#CHECKING THE MODEL USING TEST DATAS
lr_confidence=reg.score(X_test,y_test)
```

```python
#ACCURACY
print(lr_confidence)
```

```python
#PREDICTING FOR A SPECIFIC VALUE 
lr_pred=reg.predict([[144.1]])
```

```python
print(lr_pred)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,reg.predict(X_train),color='blue')
plt.title('Usage')
plt.xlabel('Quarters')
plt.ylabel("usage")
plt.show
```
